```python
# Task2 --- Build language models ---
# 1) Based on the training dataset, build unigram, bigram, 
# and trigram language models using Add-one smoothing technique.
# It is encouraged to implement models by yourself.

# Your code
# this is the code.
import re
from collections import defaultdict
def add_one_smoothing(counts, vocabulary_size):
    # Add one to all counts
    smoothed_counts = defaultdict(int)
    for key in counts:
        smoothed_counts[key] = counts[key] + 1
    
    # Add vocabulary size to denominator
    smoothed_denominator = sum(smoothed_counts.values()) + vocabulary_size
    
    return smoothed_counts, smoothed_denominator

def unigram_model(tokens):
    # Count unigrams
    unigram_counts = defaultdict(int)
    for token in tokens:
        unigram_counts[token] += 1
    
    # Add-one smoothing
    vocabulary_size = len(unigram_counts.keys())
    smoothed_unigram_counts, smoothed_unigram_denominator = add_one_smoothing(unigram_counts, vocabulary_size)
    
    # Calculate probabilities
    unigram_probs = {}
    for token in smoothed_unigram_counts:
        unigram_probs[token] = smoothed_unigram_counts[token] / smoothed_unigram_denominator
    
    return unigram_probs

def bigram_model(tokens):
    # Count bigrams
    bigram_counts = defaultdict(int)
    for i in range(len(tokens) - 1):
        bigram = (tokens[i], tokens[i+1])
        bigram_counts[bigram] += 1
    
    # Add-one smoothing
    vocabulary_size = len(set(tokens))
    smoothed_bigram_counts, smoothed_bigram_denominator = add_one_smoothing(bigram_counts, vocabulary_size)
    
    # Calculate probabilities
    bigram_probs = {}
    for bigram in smoothed_bigram_counts:
        bigram_probs[bigram] = smoothed_bigram_counts[bigram] / smoothed_bigram_denominator
    
    return bigram_probs

def trigram_model(tokens):
    # Count trigrams
    trigram_counts = defaultdict(int)
    for i in range(len(tokens) - 2):
        trigram = (tokens[i], tokens[i+1], tokens[i+2])
        trigram_counts[trigram] += 1
    
    # Add-one smoothing
    vocabulary_size = len(set(tokens))
    smoothed_trigram_counts, smoothed_trigram_denominator = add_one_smoothing(trigram_counts, vocabulary_size)
    
    # Calculate probabilities
    trigram_probs = {}
    for trigram in smoothed_trigram_counts:
        trigram_probs[trigram] = smoothed_trigram_counts[trigram] / smoothed_trigram_denominator
    
    return trigram_probs

# unigram_model = unigram_model(trainTokens)
# print("Unigram model:")
# print(unigram_model)

# bigram_model = bigram_model(trainTokens)
# print("Bigram model:")
# # print(bigram_model) # it is hard to show it in jupyter

# trigram_model = trigram_model(trainTokens)
# print("Trigram model:")
# # print(trigram_model)

# above is the model, but when I use the model, Jupyter shows the following problem that in my
# computer I cannot solve:
# IOPub data rate exceeded.
# The notebook server will temporarily stop sending output
# to the client in order to avoid crashing it.
# To change this limit, set the config variable
# `--NotebookApp.iopub_data_rate_limit`.

# Current values:
# NotebookApp.iopub_data_rate_limit=1000000.0 (bytes/sec)
# NotebookApp.rate_limit_window=3.0 (secs)

from nltk.util import ngrams
from nltk.probability import LidstoneProbDist
from nltk.tokenize import word_tokenize
trainFile = open("trainText.txt", encoding='UTF-8')
trainTxt = trainFile.read()
trainTokens = word_tokenize(trainTxt)
interpunctuations = [',', '.', ':', ';', '?', '(', ')', '[', ']', '&', '!', '*', '@', '#', '$', '%','{','}',"'",'`',"''", '``', '\\']
trainTokens = [word.lower() for word in trainTokens if word not in interpunctuations]
unigrams = ngrams(trainTokens, 1)
bigrams = ngrams(trainTokens, 2)
trigrams = ngrams(trainTokens, 3)
print(type(unigrams))
print(type(bigrams))
print(type(trigrams))
```

```python
# Task3 --- Build Naive Bayes classifiers --- 
# 1) Build Naive Bayes classifiers (with Laplace smoothing)

# Your code
import nltk
from nltk.corpus import stopwords
from nltk.classify import NaiveBayesClassifier

def extract_features(text):
    text = text.lower()
    words = nltk.word_tokenize(text)
#     words = [word for word in words if word not in stopwords.words('english')]
    features = {}
    for word in words:
        features[word] = True
    return features

resultTxt = []
with open('enwiki-train.json', 'r+') as f:
    for item in jsonlines.Reader(f):
        resultTxt.append((item['text'], item['label']))

train_features = [(extract_features(text), label) for (text, label) in resultTxt]
classifier = NaiveBayesClassifier.train(train_features, estimator=nltk.LaplaceProbDist)
print(type(classifier))
```

```python
# Task3 --- Build Naive Bayes classifiers --- 
# 2) Report Micro-F1 score and Macro-F1 score for these 
# classifiers on testing dataset explain our results.

# Your code
from sklearn.metrics import f1_score

resultTestTxt = []
with open('enwiki-test.json', 'r+') as f:
    for item in jsonlines.Reader(f):
        resultTestTxt.append((item['text'], item['label']))

true_labels = [label for (text, label) in resultTestTxt]
predicted_labels = [classifier.classify(dict(resultTestTxt)) for (text, label) in resultTestTxt]

micro_f1 = f1_score(true_labels, predicted_labels, average='micro')
macro_f1 = f1_score(true_labels, predicted_labels, average='macro')

print("Micro-F1 score: ", micro_f1)
print("Macro-F1 score: ", macro_f1)
# I think I am wrong, but I don't know the mistake 
```
